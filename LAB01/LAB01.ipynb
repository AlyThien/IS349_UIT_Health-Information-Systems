```python
import pandas as pd
import re
from datetime import datetime

def check_date(date_str):
    try:
        # Tách tháng, ngày, năm từ chuỗi mm-dd-yyyy
        year, month, day = map(int, date_str.split('-'))
        # Đảm bảo tháng và ngày có 2 chữ số, năm 4 chữ số
        formatted_date = f"{year:04d}-{month:02d}-{day:02d}"
        # Kiểm tra tính hợp lệ
        datetime.strptime(formatted_date, '%Y-%m-%d')
        return formatted_date
    except ValueError:
        return False

def check_phone(phone_str):
    patern = r'^\d{10}$'
    return bool(re.match(patern, str(phone_str)))

fields = ['MaBN', 'HoTen', 'NgaySinh', 'GioiTinh', 'DiaChi', 'SDT', 'SoBHYT', 'NgheNghiep', 'DoiTuong', 'Email', 'NguoiLienHe', 'SDT_NguoiLienHe']

n = input("Nhập số bệnh nhân bạn muốn thêm: ")
n = int(n)

new_data = {key: [] for key in fields}
for i in range(n):
    print(f"Nhập thông tin cho bệnh nhân thứ {i+1}:")
    for key in fields:
        while True:
            print(f"Nhập {key}: ")
            value = input(f"Nhập {key}: ")
            
            # Kiểm tra trống
            if value == '':
                print(f"{key} không được để trống. Vui lòng nhập lại.")
                continue
                
            # Kiểm tra định dạng ngày
            if key == 'NgaySinh':
                formatted_date = check_date(value)
                if not formatted_date:
                    print(f"Định dạng ngày không đúng. Vui lòng nhập lại (YYYY-MM-DD)")
                    continue
                value = formatted_date
                
            # Kiểm tra định dạng số điện thoại    
            if key == 'SDT' or key == 'SDT_NguoiLienHe':
                if not check_phone(value):
                    print(f"Định dạng số điện thoại không đúng. Vui lòng nhập lại (10 chữ số)")
                    continue
                    
            # Nếu mọi kiểm tra đều pass thì lưu giá trị và thoát vòng lặp
            new_data[key].append(value)
            break

# Ghi dữ liệu mới vào file CSV
df_new = pd.DataFrame(new_data)
df_new.to_csv('patients_raw.csv', mode='a', header=False, index=False)
print("Dữ liệu đã được lưu vào patients_raw.csv")
```

**Bài 2: Phát hiện lỗi**

```python
import pandas as pd
import re
from datetime import datetime

# Đọc file CSV
df = pd.read_csv('patients_raw.csv')

# Hàm kiểm tra định dạng ngày
def check_date(date_str):
    try:
        year, month, day = map(int, str(date_str).split('-'))
        formatted_date = f"{year:04d}-{month:02d}-{day:02d}"
        datetime.strptime(formatted_date, '%Y-%m-%d')
        return True
    except (ValueError, AttributeError):
        return False
    
# Hàm kiểm tra định dạng số điện thoại
def check_phone(phone_str):
    pattern = r'^\d{10}$'
    return bool(re.match(pattern, str(phone_str)))

# Hàm kiểm tra định dạng tên
def check_name(name_str):
    pattern = r'^[a-zA-Z\s]+$'
    return bool(re.match(pattern, str(name_str)))

# Hàm kiểm tra định dạng mã bệnh nhân
def check_mabn(mabn_str):
    pattern = r'^BN\d{4}$'
    return bool(re.match(pattern, str(mabn_str)))

# Hàm kiểm tra định dạng số BHYT
def check_bhyt(bhyt_str):
    pattern = r'^[A-Za-z]{2}\d{13}$'
    return bool(re.match(pattern, str(bhyt_str)))

# Khởi tạo dictionary để lưu các lỗi theo loại
error_categories = {
    'empty_fields': [],         # Các trường bị trống
    'invalid_date': [],         # Ngày sai định dạng
    'invalid_phone': [],        # SĐT sai định dạng
    'invalid_name': [],         # Tên sai định dạng
    'invalid_mabn': [],         # Mã BN sai định dạng
    'invalid_bhyt': []          # Số BHYT sai định dạng
}

# Lướt qua từng bản ghi để kiểm tra lỗi
for index, row in df.iterrows():
    # Kiểm tra trống
    empty_fields = []
    for key in df.columns:
        if pd.isna(row[key]) or str(row[key]).strip() == '':
            empty_fields.append(key)
    if empty_fields:
        error_categories['empty_fields'].append({
            'record': index + 1,
            'fields': empty_fields
        })
    
    # Kiểm tra định dạng ngày
    if not pd.isna(row['NgaySinh']) and not check_date(row['NgaySinh']):
        error_categories['invalid_date'].append({
            'record': index + 1,
            'value': row['NgaySinh']
        })
    
    # Kiểm tra định dạng số điện thoại
    if not pd.isna(row['SDT']) and not check_phone(row['SDT']):
        error_categories['invalid_phone'].append({
            'record': index + 1,
            'field': 'SDT',
            'value': row['SDT']
        })
    if not pd.isna(row['SDT_NguoiLienHe']) and not check_phone(row['SDT_NguoiLienHe']):
        error_categories['invalid_phone'].append({
            'record': index + 1,
            'field': 'SDT_NguoiLienHe',
            'value': row['SDT_NguoiLienHe']
        })
    
    # Kiểm tra định dạng tên
    if not pd.isna(row['HoTen']) and not check_name(row['HoTen']):
        error_categories['invalid_name'].append({
            'record': index + 1,
            'field': 'HoTen',
            'value': row['HoTen']
        })
    if not pd.isna(row['NguoiLienHe']) and not check_name(row['NguoiLienHe']):
        error_categories['invalid_name'].append({
            'record': index + 1,
            'field': 'NguoiLienHe',
            'value': row['NguoiLienHe']
        })
    
    # Kiểm tra định dạng mã bệnh nhân
    if not pd.isna(row['MaBN']) and not check_mabn(row['MaBN']):
        error_categories['invalid_mabn'].append({
            'record': index + 1,
            'value': row['MaBN']
        })
    
    # Kiểm tra định dạng số BHYT
    if not pd.isna(row['SoBHYT']) and not check_bhyt(row['SoBHYT']):
        error_categories['invalid_bhyt'].append({
            'record': index + 1,
            'value': row['SoBHYT']
        })

# In kết quả theo từng loại lỗi
print("\n=== PHÂN LOẠI LỖI ===")

print("\n1. Các trường bị trống:")
if not error_categories['empty_fields']:
    print("Không có trường bị trống.")
else:
    for error in error_categories['empty_fields']:
        print(f"Bản ghi {error['record']}: {', '.join(error['fields'])}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại các trường bị trống.")

print("\n2. Mã bệnh nhân sai định dạng:")
if not error_categories['invalid_mabn']:
    print("Không có mã bệnh nhân sai định dạng.")
else:
    for error in error_categories['invalid_mabn']:
        print(f"Bản ghi {error['record']}: {error['value']}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại mã bệnh nhân đúng định dạng BNxxxx (với x(4) là chữ số).")

print("\n3. Tên sai định dạng:")
if not error_categories['invalid_name']:
    print("Không có tên sai định dạng.")
else:
    for error in error_categories['invalid_name']:
        print(f"Bản ghi {error['record']} - {error['field']}: {error['value']}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại tên chỉ gồm chữ cái và khoảng trắng.")

print("\n4. Ngày sinh sai định dạng:")
if not error_categories['invalid_date']:
    print("Không có ngày sinh sai định dạng.")
else:
    for error in error_categories['invalid_date']:
        print(f"Bản ghi {error['record']}: {error['value']}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại ngày sinh đúng định dạng YYYY-MM-DD.")

print("\n5. Số điện thoại sai định dạng:")
if not error_categories['invalid_phone']:
    print("Không có số điện thoại sai định dạng.")
else:
    for error in error_categories['invalid_phone']:
        print(f"Bản ghi {error['record']} - {error['field']}: {error['value']}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại số điện thoại đúng định dạng 10 chữ số.")

print("\n6. Số BHYT sai định dạng:")
if not error_categories['invalid_bhyt']:
    print("Không có số BHYT sai định dạng.")
else:
    for error in error_categories['invalid_bhyt']:
        print(f"Bản ghi {error['record']}: {error['value']}")
    print("\n==> Giải quyết lỗi bằng cách nhập lại số BHYT đúng định dạng XXyyyyyyyyyyyyy (với X(2) là ký tự và y(13) là chữ số).")

    print("\n=== GIẢI QUYẾT LỖI ===")
while error_categories:
    if error_categories['invalid_mabn']:
        print("\n2. Mã bệnh nhân sai định dạng:")
        for error in error_categories['invalid_mabn']:
            print(f"Sửa mã bệnh nhân của bản ghi {error['record']}:  ")
            new_value = input()
            while not check_mabn(new_value):
                print("Mã bệnh nhân không đúng định dạng BNxxxx (với x(4) là chữ số). Vui lòng nhập lại: ")
                new_value = input()
            df.at[error['record'] - 1, 'MaBN'] = new_value
        error_categories['invalid_mabn'] = []
    elif error_categories['invalid_name']:
        print("\n3. Tên sai định dạng:")
        for error in error_categories['invalid_name']:
            print(f"Sửa tên của bản ghi {error['record']} - {error['field']}:  ")
            new_value = input()
            while not check_name(new_value):
                print("Tên chỉ gồm chữ cái và khoảng trắng. Vui lòng nhập lại: ")
                new_value = input()
            df.at[error['record'] - 1, error['field']] = new_value
        error_categories['invalid_name'] = []
    elif error_categories['invalid_date']:
        print("\n4. Ngày sinh sai định dạng:")
        for error in error_categories['invalid_date']:
            print(f"Sửa ngày sinh của bản ghi {error['record']}:  ")
            while not check_date(new_value):
                print("Định dạng ngày không đúng. Vui lòng nhập lại (YYYY-MM-DD): ")
                new_value = input()
            df.at[error['record'] - 1, 'NgaySinh'] = new_value
        error_categories['invalid_date'] = []
    elif error_categories['invalid_phone']:
        print("\n5. Số điện thoại sai định dạng:")
        for error in error_categories['invalid_phone']:
            print(f"Sửa số điện thoại của bản ghi {error['record']} - {error['field']}:  ")
            new_value = input()
            while not check_phone(new_value):
                print("Định dạng số điện thoại không đúng. Vui lòng nhập lại (10 chữ số): ")
                new_value = input()
            df.at[error['record'] - 1, error['field']] = new_value
        error_categories['invalid_phone'] = []
    elif error_categories['invalid_bhyt']:
        print("\n6. Số BHYT sai định dạng:")
        for error in error_categories['invalid_bhyt']:
            print(f"Sửa số BHYT của bản ghi {error['record']}:  ")
            new_value = input()
            while not check_bhyt(new_value):
                print("Định dạng số BHYT không đúng. Vui lòng nhập lại (XXyyyyyyyyyyyyy với X(2) là ký tự và y(13) là chữ số): ")
                new_value = input()
            df.at[error['record'] - 1, 'SoBHYT'] = new_value
        error_categories['invalid_bhyt'] = []
    else:
        break
```

**Bài 3: Chuẩn hóa dữ liệu**

```python
import pandas as pd
import re
from datetime import datetime

# Đọc file CSV
df = pd.read_csv('patients_raw.csv')

# Hàm chuẩn hóa giới tính
def normalize_gender(gender_str):
    gender_str = str(gender_str).lower().strip()
    if gender_str in ['nam', 'm', 'male', 'Ná»¯', 'M']:
        return 'male'
    elif gender_str in ['nu', 'f', 'female', 'F']:
        return 'female'
    else:
        return 'other'  

# Hàm chuẩn hóa địa chỉ Quận/Huyện
    # Tách DiaChi thành 3 phần dựa trên dấu phẩy
df[["DiaChiNha", "Quan", "TinhThanh"]] = df["DiaChi"].str.split(",", n=2, expand=True)

# Xóa khoảng trắng dư thừa
df["DiaChiNha"] = df["DiaChiNha"].str.strip()
df["Quan"] = df["Quan"].str.strip()
df["TinhThanh"] = df["TinhThanh"].str.strip()

    # Hàm chuẩn hóa Quận/Huyện
def normalize_quan(text):
    if pd.isna(text): 
        return text
    
    text = str(text).lower().strip()
    
    # Chuyển đổi district sang quận
    text = text.replace('district', 'quan')
    
    # Loại bỏ viết tắt và chuẩn hóa
    text = re.sub(r"^q\s*([0-9]+)$", r"quan \1", text)  # Q1 -> quan 1
    text = re.sub(r"^d\s*([0-9]+)$", r"quan \1", text)  # D1 -> quan 1
    text = text.replace("quan", "quận")
    
    # Viết hoa chữ cái đầu
    text = text.title()
    return text

df["Quan"] = df["Quan"].apply(normalize_quan)

mapping_tinh = {
        # Hồ Chí Minh
    "hcmc": "Thành phố Hồ Chí Minh",
    "tp.hcm": "Thành phố Hồ Chí Minh",
    "tphcm": "Thành phố Hồ Chí Minh",
    "ho chi minh": "Thành phố Hồ Chí Minh",
    "sai gon": "Thành phố Hồ Chí Minh",
    "sg": "Thành phố Hồ Chí Minh",

        # Hà Nội
    "hn": "Thành phố Hà Nội",
    "ha noi": "Thành phố Hà Nội",
    "hanoi": "Thành phố Hà Nội",
    "tp.hn": "Thành phố Hà Nội",

        # Đà Nẵng
    "dn": "Thành phố Đà Nẵng",
    "da nang": "Thành phố Đà Nẵng",
    "tpdn": "Thành phố Đà Nẵng",

        # Cần Thơ
    "can tho": "Thành phố Cần Thơ",
    "ct": "Thành phố Cần Thơ",

        # Huế
    "hue": "Thành phố Huế (Thừa Thiên Huế)",
    "thua thien hue": "Thành phố Huế (Thừa Thiên Huế)",
}

# hàm chuẩn hóa 
def normalize_tinh(text):
    if pd.isna(text):
        return text
    
    t = str(text).lower().strip()
    
    # Nếu trong mapping viết tắt/thông dụng
    if t in mapping_tinh:
        return mapping_tinh[t]
    
    # Viết hoa chữ cái đầu cho trường hợp đã đúng tên
    return text.title()

# Chuẩn hóa TinhThanh
df["TinhThanh"] = df["TinhThanh"].apply(normalize_tinh)

# Hàm chuẩn hóa BHYT (2 chữ cái + 13 số)
def normalize_bhyt(bhyt_str):
    bhyt_str = str(bhyt_str).strip().upper()  # Chuyển thành chữ hoa
    match = re.match(r'^([A-Za-z]{2})(\d{13})$', bhyt_str)
    if match:
        return match.group(1) + match.group(2)  # Giữ nguyên 2 chữ cái + 13 số
    return bhyt_str  # Nếu không khớp, giữ nguyên (có thể báo lỗi sau)

# Áp dụng chuẩn hóa
df['GioiTinh'] = df['GioiTinh'].apply(normalize_gender)
df["TinhThanh"] = df["TinhThanh"].apply(normalize_tinh) # Chuẩn hóa TinhThanh
df['SoBHYT'] = df['SoBHYT'].apply(normalize_bhyt)

# Ghi lại file CSV đã chuẩn hóa (tạo file mới để giữ nguyên file cũ)
df.to_csv('patients_normalized.csv', index=False)
print("Dữ liệu đã được chuẩn hóa và lưu vào 'patients_normalized.csv'!")
```

**Bài 4-5: Trùng lặp và privacy**

```python
import pandas as pd
import hashlib

# Đọc file CSV (em thay tên file nếu cần)
df = pd.read_csv('patients_raw.csv')

# PHÁT HIỆN TRÙNG LẶP
# Kiểm tra trùng lặp dựa trên các trường chính: HoTen, NgaySinh, SDT
duplicates = df[df.duplicated(subset=['HoTen', 'NgaySinh', 'SDT'], keep=False)]

if not duplicates.empty:
    print("Phát hiện các bệnh nhân trùng lặp:")
    print(duplicates)
    # Gợi ý: Xóa trùng lặp nếu cần (giữ dòng đầu tiên)
    df_clean = df.drop_duplicates(subset=['HoTen', 'NgaySinh', 'SDT'], keep='first')
    print(f"Đã loại bỏ {len(df) - len(df_clean)} dòng trùng lặp.")
else:
    print("Không phát hiện trùng lặp trong dữ liệu.")
    df_clean = df.copy()

# PRIVACY
# Hàm mã hóa để ẩn danh (sử dụng SHA-256 cho an toàn)
def anonymize(value):
    if pd.isna(value):
        return 'Anonymous'
    return hashlib.sha256(str(value).encode()).hexdigest()[:10]  # Lấy 10 ký tự đầu để ngắn gọn

# Các trường cần ẩn danh
anon_columns = ['HoTen', 'DiaChi', 'SDT', 'Email', 'NguoiLienHe', 'SDT_NguoiLienHe']

# Tạo DataFrame ẩn danh
df_anon = df_clean.copy()
for col in anon_columns:
    if col in df_anon.columns:
        df_anon[col] = df_anon[col].apply(anonymize)

# Lưu file ẩn danh mới
df_anon.to_csv('patients_anonymous.csv', index=False)
print("Phiên bản ẩn danh đã được tạo và lưu vào 'patients_anonymous.csv'!")

# In mẫu để kiểm tra
print("\nMẫu dữ liệu ẩn danh:")
df_anon.head()
```
